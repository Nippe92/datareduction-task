# Sleep patterns of railroad employees

# About the dataset
Link: https://catalog.data.gov/dataset/work-schedules-and-sleep-patterns-of-railroad-employees

# Description
The Federal Railroad Administration (FRA) sponsored a study of the work schedules and sleep patterns of railroad employees. The purpose of the study was to understand work-schedule related fatigue that affects various categories of railroad employees by documenting a group's work/rest schedules and sleep patterns to ascertain their impact on the level of fatigue/alertness. Employees surveyed include: signalmen, maintenance of way (MOW) workers, dispatchers, and train & engine service workers (in both freight and passenger train service)

## CA - Age group vs Job pressure

```python
import pandas as pd
import matplotlib.pyplot as plt
import prince

contingency_table = pd.crosstab(
    [railroad_df_clean['Age_Group'], railroad_df_clean['Sex']],
    railroad_df_clean['Job_pressure'])

ca = prince.CA(n_components=2, n_iter=10)
ca = ca.fit(contingency_table)

row_coordinates = ca.row_coordinates(contingency_table)
column_coordinates = ca.column_coordinates(contingency_table)

age_group_labels = {
    1: '20-29', 
    2: '30-39', 
    3: '40-49', 
    4: '50-59', 
    5: '60+'
}

gender_labels = {
    1: 'Male', 
    2: 'Female', 
}

job_pressure_labels = {
    1: 'No stress', 
    2: 'A little stress', 
    3: 'Stressful', 
    4: 'Very stressful'
}

plt.figure(figsize=(18, 12))
plt.scatter(row_coordinates[0], row_coordinates[1], label='Age Group', c='blue', s=100)
plt.scatter(column_coordinates[0], column_coordinates[1], label='Job Pressure', c='red', s=100)

for i, txt in enumerate(contingency_table.index):
    age_group, gender = txt
    combined_label = f"{age_group_labels.get(age_group, age_group)} ({gender_labels.get(gender, gender)})"
    plt.annotate(
        combined_label,
        (row_coordinates.iloc[i, 0] + 0.02, row_coordinates.iloc[i, 1]),
        color='blue'
    )

for i, txt in enumerate(contingency_table.columns):
    pressure_label = job_pressure_labels.get(txt, f'Pressure {txt}')
    plt.annotate(
        pressure_label,
        (column_coordinates.iloc[i, 0] + 0.02, column_coordinates.iloc[i, 1]),
        color='red'
    )

explained_variance = ca.percentage_of_variance_
cumulative_explained_variance = ca.cumulative_percentage_of_variance_

print("Explained Variance for each component (%): ")
print(explained_variance)
print("Cumulative Explained Variance for each component (%): ")
print(cumulative_explained_variance)

plt.axhline(0, color='black', linestyle='--', linewidth=1.2)
plt.axvline(0, color='black', linestyle='--', linewidth=1.2)
plt.scatter(0, 0, color='black', s=50, zorder=5)

plt.xlabel(f'Component 1 {explained_variance[0]: .2f}%')
plt.ylabel(f'Component 2 {explained_variance[1]: .2f}%')
plt.title('Correspondence Analysis: Age_Group vs Job Pressure')
plt.legend()
plt.grid(True)
plt.show()

print("Eigenvalues:", ca.eigenvalues_)
```

```python
import numpy as np

P = contingency_table / contingency_table.values.sum()

r = P.sum(axis=1).values.reshape(-1, 1)
c = P.sum(axis=0).values.reshape(1, -1)

expected = r @ c

S_std = (P - expected) / np.sqrt(expected)

U, s, Vt = np.linalg.svd(S_std, full_matrices=False)


plt.plot(s, marker="o")
plt.title("Singular Values (Scree plot)")
plt.xlabel("Index")
plt.ylabel("Singular Value")
plt.show()
```

Test to see the distrobution between having- and not having children

```python
from scipy.stats import mannwhitneyu

with_children = railroad_df_clean[railroad_df_clean['Childrendependents'] > 0]['Childrendependents']
without_children = railroad_df_clean[railroad_df_clean['Childrendependents'] == 0]['Childrendependents']

stat, p_value = mannwhitneyu(with_children, without_children, alternative='two-sided')

print(f"Mann-Whitney U-statistik: {stat}")
print(f"P-värde: {p_value}")

if p_value < 0.05:
    print("Det finns en signifikant skillnad mellan grupperna (p < 0.05).")
else:
    print("Det finns ingen signifikant skillnad mellan grupperna (p >= 0.05).")
```

```python

railroad_df_clean['Has_Children'] = railroad_df_clean['Childrendependents'] > 0

railroad_df_clean['Has_Children'] = railroad_df_clean['Has_Children'].map({True: 'Have Children', False: 'No Children'})

def group_age(age_group):
    if age_group in [1, 2]:  # 20-39
        return '20-39'
    elif age_group in [3, 4]:  # 40-59
        return '40-59'
    elif age_group == 5:  # 60+
        return '60+'
    else:
        return 'Unknown'

railroad_df_clean['Age_Group_Grouped'] = railroad_df_clean['Age_Group'].apply(group_age)

print(railroad_df_clean['Age_Group_Grouped'].value_counts())

railroad_df_clean.count()

contingency_table = pd.crosstab(
    [railroad_df_clean['Age_Group_Grouped'], railroad_df_clean['Has_Children'] ],
    railroad_df_clean['Sleep_loss'])

print(contingency_table)
```

```python
ca = prince.CA(n_components=2, n_iter=10)
ca = ca.fit(contingency_table)

row_coordinates = ca.row_coordinates(contingency_table)
column_coordinates = ca.column_coordinates(contingency_table)


sleep_loss_labels = {
    1: 'No stress', 
    2: 'A little stress', 
    3: 'Stressful', 
    4: 'Very stressful'
}

plt.figure(figsize=(18, 12))
plt.scatter(row_coordinates[0], row_coordinates[1], label='employee_groups', c='blue', s=100)
plt.scatter(column_coordinates[0], column_coordinates[1], label='Sleep_loss', c='red', s=100)


for i, txt in enumerate(contingency_table.index):
    Age_Group_Grouped, Has_Children = txt
    combined_label = f"{Age_Group_Grouped} - {Has_Children} "
    plt.annotate(
        combined_label,
        (row_coordinates.iloc[i, 0] + 0.02, row_coordinates.iloc[i, 1]),  # Offset x-coordinate
        color='blue'
    )

for i, txt in enumerate(contingency_table.columns):
    sleep_loss_label = sleep_loss_labels.get(txt, f'Sleep_loss {txt}')
    plt.annotate(
        sleep_loss_label,
        (column_coordinates.iloc[i, 0] + 0.02, column_coordinates.iloc[i, 1]),  # Offset x-coordinate
        color='red'
    )
    
explained_variance = ca.percentage_of_variance_
cumulative_explained_variance = ca.cumulative_percentage_of_variance_

print("Explained Variance for each component (%): ")
print(explained_variance)
print("Cumulative Explained Variance for each component (%): ")
print(cumulative_explained_variance)

plt.axhline(0, color='black', linestyle='--', linewidth=1.2)
plt.axvline(0, color='black', linestyle='--', linewidth=1.2)
plt.scatter(0, 0, color='black', s=50, zorder=5)

plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.xlabel(f'Component 1 {explained_variance[0]: .2f}%')
plt.ylabel(f'Component 2 {explained_variance[1]: .2f}%')
plt.title('Correspondence Analysis: Age Groups and parental status vs Sleep loss')
plt.legend()
plt.grid(True)
plt.show()

print("Eigenvalues:", ca.eigenvalues_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import prince

bins = [0, 2, 5, float('inf')]
labels = ['Low consumption', 'Normal consumption', 'High consumption']

railroad_df_clean['Caffeine_Consumption_Level'] = pd.cut(
    railroad_df_clean['If_yes_how_many_daily'], 
    bins=bins, 
    labels=labels, 
    right=False
)

contingency_table = pd.crosstab(
    railroad_df_clean['Caffeine_Consumption_Level'], 
    railroad_df_clean['Sleep_loss']
)

sleep_loss_labels = {
    1: 'No stress', 
    2: 'A little stress', 
    3: 'Stressful', 
    4: 'Very stressful'
}
contingency_table.columns = contingency_table.columns.map(sleep_loss_labels)

ca = prince.CA(n_components=2, n_iter=10)
ca = ca.fit(contingency_table)

row_coordinates = ca.row_coordinates(contingency_table)
column_coordinates = ca.column_coordinates(contingency_table)

plt.figure(figsize=(10, 8))

plt.scatter(row_coordinates[0], row_coordinates[1], label='Caffeine Consumption', c='red', s=50)
for i, txt in enumerate(contingency_table.index):
    plt.annotate(txt, (row_coordinates.iloc[i, 0], row_coordinates.iloc[i, 1] + 0.005), color='red', ha='center')

plt.scatter(column_coordinates[0], column_coordinates[1], label='Sleep Loss', c='blue', s=50)
for i, txt in enumerate(contingency_table.columns):
    sleep_loss_label = sleep_loss_labels.get(txt, str(txt))
    plt.annotate(
        sleep_loss_label,
        (column_coordinates.iloc[i, 0], column_coordinates.iloc[i, 1] + 0.005),
        color='blue',
        ha='center',  
        fontsize=10  
    )
    
    
explained_variance = ca.percentage_of_variance_
cumulative_explained_variance = ca.cumulative_percentage_of_variance_

print("Explained Variance for each component (%): ")
print(explained_variance)
print("Cumulative Explained Variance for each component (%): ")
print(f"{cumulative_explained_variance}")
print("Eigenvalues:", ca.eigenvalues_)

plt.xlim(-0.1, 0.3)  
plt.ylim(-0.05, 0.15)  

plt.axhline(0, color='black', linestyle='--', linewidth=1.2)
plt.axvline(0, color='black', linestyle='--', linewidth=1.2)
plt.scatter(0, 0, color='black', s=50, zorder=5)

plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.title('Correspondence Analysis: Caffeine Consumption vs Sleep Loss')
plt.xlabel(f'Component 1 {explained_variance[0]: .2f}%')
plt.ylabel(f'Component 2 {explained_variance[1]: .2f}%')
plt.legend()
plt.grid(True)
plt.show()
```

```python
P = contingency_table / contingency_table.values.sum()

r = P.sum(axis=1).values.reshape(-1, 1)
c = P.sum(axis=0).values.reshape(1, -1)

expected = r @ c

S_std = (P - expected) / np.sqrt(expected)

U, s, Vt = np.linalg.svd(S_std, full_matrices=False)


plt.plot(s, marker="o")
plt.title("Singular Values (Scree plot)")
plt.xlabel("Index")
plt.ylabel("Singular Value")
plt.show()
```

Moderate consumers fall somewhere in between, not strongly pulled towards any extreme.
