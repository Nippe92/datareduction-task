# Sleep patterns of railroad employees

# About the dataset
Link: https://catalog.data.gov/dataset/work-schedules-and-sleep-patterns-of-railroad-employees

# Description
The Federal Railroad Administration (FRA) sponsored a study of the work schedules and sleep patterns of railroad employees. The purpose of the study was to understand work-schedule related fatigue that affects various categories of railroad employees by documenting a group's work/rest schedules and sleep patterns to ascertain their impact on the level of fatigue/alertness. Employees surveyed include: signalmen, maintenance of way (MOW) workers, dispatchers, and train & engine service workers (in both freight and passenger train service)

### CA TEST - JOB TYPE VS JOB  PRESSURE

```python
import pandas as pd
import matplotlib.pyplot as plt
import prince

# Step 1: Create the contingency table
contingency_table = pd.crosstab(
    [railroad_df['Age_Group'], railroad_df['Sex']],
    railroad_df['Job_pressure'])

# Step 2: Initialize and fit the CA model
ca = prince.CA(n_components=2, n_iter=10)
ca = ca.fit(contingency_table)

# Step 3: Get row and column coordinates
row_coordinates = ca.row_coordinates(contingency_table)
column_coordinates = ca.column_coordinates(contingency_table)

# Step 4: Define custom labels
age_group_labels = {
    1: '20-29', 
    2: '30-39', 
    3: '40-49', 
    4: '50-59', 
    5: '60+'
}

gender_labels = {
    1: 'Male', 
    2: 'Female', 
}

job_pressure_labels = {
    1: 'No stress', 
    2: 'A little stress', 
    3: 'Stressful', 
    4: 'Very stressful'
}

# Step 5: Plot the CA results
plt.figure(figsize=(10, 7))
plt.scatter(row_coordinates[0], row_coordinates[1], label='Age_Group', c='blue', s=100)
plt.scatter(column_coordinates[0], column_coordinates[1], label='Job_pressure', c='red', s=100)

# Annotate Age Group and Gender
for i, txt in enumerate(contingency_table.index):
    # Extract Age Group and Gender from the multi-index
    age_group, gender = txt
    combined_label = f"{age_group_labels.get(age_group, age_group)} {gender_labels.get(gender, gender)}"
    plt.annotate(combined_label, (row_coordinates.iloc[i, 0], row_coordinates.iloc[i, 1]), color='blue')

# Annotate Job Pressure
for i, txt in enumerate(contingency_table.columns):
    pressure_label = job_pressure_labels.get(txt, f'Pressure {txt}')
    plt.annotate(pressure_label, (column_coordinates.iloc[i, 0], column_coordinates.iloc[i, 1]), color='red')

# Include explained variance for each component
explained_variance = ca.percentage_of_variance_
cumulative_explained_variance = ca.cumulative_percentage_of_variance_

print("Explained Variance for each component (%): ")
print(explained_variance)
print("Cumulative Explained Variance for each component (%): ")
print(cumulative_explained_variance)

# Axis and grid
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.xlabel(f'Component 1 {explained_variance[0]: .2f}%')
plt.ylabel(f'Component 2 {explained_variance[1]: .2f}%')
plt.title('Correspondence Analysis: Age_Group vs Job Pressure')
plt.legend()
plt.grid(True)
plt.show()

# Optional: Check explained variance
print("Eigenvalues:", ca.eigenvalues_)
```

### Analysis


```python
print("Explained Variance for each component (%): ")
print(explained_variance)
print("Cumulative Explained Variance for each component (%): ")
print(cumulative_explained_variance)
```

```python
import numpy as np

# Step 1: Relative frequencies
P = contingency_table / contingency_table.values.sum()

# Step 2: Row and column sums
r = P.sum(axis=1).values.reshape(-1, 1)
c = P.sum(axis=0).values.reshape(1, -1)

# Step 3: Expected frequencies under independence
expected = r @ c

# Step 4: Standardized residuals
S_std = (P - expected) / np.sqrt(expected)

# Step 5: SVD
U, s, Vt = np.linalg.svd(S_std, full_matrices=False)

# Singular values (s), left singular vectors (U), right singular vectors (Vt)

plt.plot(s, marker="o")
plt.title("Singular Values (Scree plot)")
plt.xlabel("Index")
plt.ylabel("Singular Value")
plt.show()
```
