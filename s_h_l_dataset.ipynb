```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shl_data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
shl_data.info()
```

Datacleaning

```python
#check for duplicates
duplicates = shl_data.duplicated().sum()
print(duplicates)
```

```python
# Check for missing values
missing_values = shl_data.isnull().sum()
print(missing_values)
#found 219 missing values on sleep disorder column
```

Noticed that the None value should be there after research. The meaning of None is that the person in question doesnt have sleep disorder

```python
#made None to a category we can use
none_count = (shl_data['Sleep Disorder'] == 'None').sum()
print(f"Amount of 'None': {none_count}")
shl_data['Sleep Disorder'].fillna('None', inplace=True)
missing_values = shl_data['Sleep Disorder'].isnull().sum()
print(f"missing values after cleaning: {missing_values}")
```

```python
#Checking the object columns to see that there is no double where there might be a small/large letter difference
print(shl_data['Gender'].unique())
print(shl_data['Gender'].value_counts())
print(shl_data['Occupation'].unique())
print(shl_data['Occupation'].value_counts())
print(shl_data['BMI Category'].unique())
print(shl_data['BMI Category'].value_counts())
print(shl_data['Blood Pressure'].unique())
print(shl_data['Blood Pressure'].value_counts())
print(shl_data['Sleep Disorder'].unique())
print(shl_data['Sleep Disorder'].value_counts())
```
